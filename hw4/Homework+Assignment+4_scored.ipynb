Let $Q$ be a real, square orthogonal matrix. Show that all eigenvalues of $Q$ satisfy $|\lambda |=1$.

If $Q$ is orthogonal then, 

\begin{align*}
Q^\top Q &= I \\
\det(Q^\top Q) &= \det(I) \\
\end{align*}

$${\displaystyle I = \left( \begin{array}{rrrrr} 1 &  0  &  0 &  \ldots \\ 0  & 1 & 0 & \ldots   \\ 0 & 0 & 1 & \ldots \\ \vdots &\vdots &\vdots & \ddots \end{array} \right) \,}$$

Determinant of identity matrix is 1 since all the diagonal elements are 1. Moreover, the diagonals of a 'diagonal' matrix are the eigen values of that respective matrix. $\therefore |\lambda| = 1$.

# (b)
A projection matrix is a matrix $P$ for which $P^2=P$. Find the eigenvalues of a projection matrix.

Let $P$ be the projection matrix where $\lambda$ is the set of eigenvalues and $v$ is the eigen vector.

We know,

\begin{align*}
Pv &= \lambda v \hspace{3em}\ldots(1)  \\
\text{Multiplying both sides with }P \\
P^2 v &= P \lambda v \\
P^2 v &= \lambda P v \\
P v &= \lambda (\lambda v) \hspace{2.5em}\ldots \text{from 1 and definition of projection matrix} \\
P v &= \lambda^2 v \hspace{3em}\ldots(2)  \\
\text{From (1) and (2), } \\
\lambda^2 v - \lambda v &= 0 \\
v(\lambda^2 - \lambda) &= 0 \\
\lambda(\lambda -1)v &= 0\\
\Rightarrow \lambda = 0, 1 & \hspace{.5em}\text{ when } \hspace{.5em}v \neq 0
\end{align*}

$\therefore$ 0, 1 are the eigenvalues of the projection matrix

Let $A$ be a $n × n$ tridiagonal matrix that results (up to an overall constant scaling factor) from an
$n$-point spatial discretization of the Laplace operator in $1$ spatial dimension. Specifically, all diagonal
elements of $A$ have the value $-2$, all elements just below or above the diagonal have value $1$.

$${\displaystyle A = \left( \begin{array}{rrrrr} -2 &  1  &  0 &  0& \ldots \\ 1  & -2 & 1 & 0 &\ldots   \\ 0 & 1 & -2 &  1 &\ldots \\ \vdots & & & & \ddots \end{array} \right) \,}$$



Take $n = 150$. Use `scipy.linalg.qr` to implement Algorithm 4.7 from the book and compute the full eigenvalue spectrum of $A$. Compare against
the eigenvalues obtained by the built-in routine `scipy.linalg.eig(A)`. Show in a figure the (approximate)
eigenvalues after 10, 100 and 500 QR iterations, as well as the eigenvalues obtained with `eig(A)`.


```python
def qr_iter(N, lim, prec):
    
    '''
    N : Matrix dimension
    lim: Number of QR iterations
    prec: Precision (approximation) to mentioned number of digits
    '''
    
    main_dia = np.eye(N)*-2
    diagonals = [np.diagonal(main_dia), [1], [1]]
    A = diags(diagonals, [0, 1, -1]).toarray()

    lamda, _ = linalg.eig(A)
    k = 0

    lamda_real = np.round(lamda.real, prec)
    diag_A = np.round(np.diagonal(A), prec)

    while(k < lim):
        q, r = linalg.qr(A)
        A = r @ q

        diag_A = np.round(np.diagonal(A), prec)
        k = k+1
    
    ei, _ = linalg.eig(A)
    
    return lamda_real, ei.real
```

```python
l_real, ei_s_k10 = qr_iter(150, 10, 4)
_, ei_s_k100 = qr_iter(150, 100, 4)
_, ei_s_k500 = qr_iter(150, 500, 4)
```

```python
plt.rcParams["figure.figsize"] = (16, 12)
plt.plot(l_real, "#914040", label = "eigen values with A", alpha = 0.5)
plt.plot(ei_s_k10, "#4A3FC4", label = "qr-iter eigen values @ k = 10", linewidth = 8)
plt.plot(ei_s_k100, "#73FF3F", label = "qr-iter eigen values @ k = 100", linewidth = 6)
plt.plot(ei_s_k500, "#F39641", label = "qr-iter eigen values @ k = 500", linewidth = 3)
plt.xlabel(r'$i^{th}$ Eigenvalue', fontsize = 14)
plt.ylabel(r'Eigenvalue', fontsize = 14)
plt.title(r'Eigen values (only real) after $k$ QR-iterations on 150 x 150 tridiagonal matrix', fontsize = 14)
plt.legend()
plt.show() 
```

It can be observed that increasing QR iterations smoothens the eigenvalues.
