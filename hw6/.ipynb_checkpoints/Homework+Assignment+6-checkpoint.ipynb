# Exercise 1
Let $P_k$ be the $k$th _Legendre polynomial_ (see paragraph 7.3.4) on $[-1, 1]$.

We know that $P_0(x) = 1$, $P_1(x) = x$, and that $P_k$ satisfy the recurrence relation 

$$k P_k(x) =x (2k-1)  P_{k-1}(x) - (k-1) P_{k-2}(x).$$

We will use [_induction_](https://en.wikipedia.org/wiki/Mathematical_induction) to show that the derivatives of the Legendre polynomials satisfy

$$P_k'(1) = k(k+1)/2.$$

## (a) Base case
Show that $P_0'(1) = 0$ and $P_1'(1) = 1$.

At k = 0,

$$
P_0(x) = 1 \hspace{3em}\text{(given)} \\
\text{differentiate w.r.t } x \text{ and }x = 1 \\
P_0'(1) = 0 \\
% \Rightarrow P_0'(1) = 0(0+1)/2 = 0(0.5) = 0 \\
$$

At k = 1,

$$
1 \cdot P_1(x) =x (2(1)-1)  P_{1-1}(x) - (1-1) P_{1-2}(x) \\
\rightarrow P_1(x) = x (1) P_0(x) \\
\text{differentiate w.r.t } x \text{ and } x = 1\\
\rightarrow P_1'(1) = 1 (1) (1) \\
\Rightarrow P_1'(1) = 1 \\
% P_1'(1) = 1(1+1)/2 = 1(1) = 1
$$

Hence the base cases are proved


## (b)  Induction Step
Let $k \geq 2$. Use the recurrence relation to show that $P_k'(1) = k(k+1)/2$.

Pg 322

Assume, $P_k(x)$ is true for k = n. 

$$ \Rightarrow nP_n(x) = x (2n-1)  P_{n-1}(x) - (n-1) P_{n-2}(x) \hspace{2em}\text{is true}$$

We will prove by induction that k = n-1 also satisfies the relation for its derivative.

At k = 2

$$ 
\Rightarrow 2\cdot P_{2}(x) = x (2(2)-1)  P_{1}(x) - (1) P_{0}(x) \\
2\cdot P_{2}(x) = x (3)  P_{1}(x) - P_{0}(x) \\
2\cdot P_{2}'(x) = 3 (P_{1}(x) + xP_{1}'(x) )- P_{0}'(x) \\ 
$$

$$
2\cdot P_{2}'(1) = 3 (P_{1}(1) + 1 \cdot P_{1}'(1) )- P_{0}'(1) \\
2\cdot P_{2}'(1) = 3 (P_{1}(1) + 1 \cdot P_{1}'(1) )- P_{0}'(1) \\
2\cdot P_{2}'(1) = 3 (1 + 1 \cdot 1 )- 0 \\
P_{2}'(1) = 3  \\
$$


At k = 3

$$ 
\Rightarrow 3\cdot P_{3}(x) = x (2(3)-1)  P_{2}(x) - (2) P_{1}(x) \\
3\cdot P_{3}(x) = x (5)  P_{2}(x) - 2 \cdot P_{1}(x) \\
3\cdot P_{3}'(x) = 5 (P_{2}(x) + xP_{2}'(x) ) - 2 \cdot P_{1}'(x) \\ 
$$

$$
3\cdot P_{3}'(1) = 5 (P_{2}(1) + (1)P_{2}'(1) ) - 2 \cdot P_{1}'(1) \\ 
3\cdot P_{3}'(1) = 5 (1 + 3 ) - 2 \cdot 1 \\ 
3\cdot P_{3}'(1) = 5 (4)- 2 \\
3\cdot P_{3}'(1) = 18 \\
P_{3}'(1) = 18/3 = 6  \\
$$

$\vdots$

At k = n-1

$$ 
\Rightarrow (n-1)\cdot P_{n-1}(x) = x (2(n-1)-1)  P_{n-2}(x) - (n-2) P_{n-3}(x) \\
\Rightarrow (n-1)\cdot P_{n-1}(x) = x (2n-3)  P_{n-2}(x) - (n-2) P_{n-3}(x) \\
\Rightarrow (n-1)\cdot P_{n-1}'(x) = (2n-3)(P_{n-2}(x) + x \cdot P_{n-2}'(x)) - (n-2) P_{n-3}'(x) \\
$$

$$
\Rightarrow (n-1)\cdot P_{n-1}'(1) = (2n-3)(P_{n-2}(1) + 1 \cdot P_{n-2}'(1)) - (n-2) P_{n-3}'(1) \\
\Rightarrow (n-1)\cdot P_{n-1}'(1) = (2n-3)(1 + \cfrac{(n-2)(n-1)}{2}) - (n-2)\cfrac{(n-3)(n-2)}{2} \\
\Rightarrow P_{n-1}'(1) = \cfrac{(2n-3)(n^2-3n+4) - (n-2)^2 (n-3)}{2(n-1)}\\
\Rightarrow P_{n-1}'(1) = \cfrac{n^3 - 2n^2 - n}{2(n-1)}\\
\Rightarrow P_{n-1}'(1) = \cfrac{n(n-1)^2}{2(n-1)}\\
\Rightarrow P_{n-1}'(1) = \cfrac{n(n-1)}{2}\\
$$

Since the result is valid till k = n-1, it will also remain the same for k = n. i.e.  $P_k'(1) = k(k+1)/2$.

Hence  $P_k'(1) = k(k+1)/2$ is proved.

# (a)
Read chapter 6.6 on Nonlinear Least squares. Use the Gauss-Newton Method to fit the model function $f(t, x_1, x_2) = x_1\!\cdot x_2^t$ to the data. Find estimates for the initial population $P_0=x_1$ and the growth rate $r=x_2$.

```python
def f(t, x):
    '''
    t: vector of time points
    x: x vector
    '''
#     return x[0]*np.exp(x[1]*t)
    return x[0] * x[1] ** t

def log_f(t, x):
    '''
    t: vector of time points
    x: x vector
    '''
    return t*np.log(x[0]) + t*np.log(x[1])    
#     return np.log(x[0]) + t*x[1]
```

```python
def r(y, f, t, x):
    '''
    y : true data points
    f : pass the true function or log function
    Computes residual
    '''
    return y - f(t, x)

def J(t, x, log = 0):
    '''
    Jacobian of the residual
    log: T/F : boolean
    '''
    J = np.zeros((len(t), 2))

    for i in range(J.shape[0]):
        
        if log:
            J[i][0] = -1*(t[i]/x[0] + t[i]*np.log(x[1]))            
            J[i][1] = -1*(t[i]/x[1] + t[i]*np.log(x[0]))
            
#             J[i][0] = -1*(1/x[0])            
#             J[i][1] = -1*(t[i])
            
        else:
            J[i][0] = -x[1]**t[i]            
            J[i][1] = -t[i]*x[0]*x[1]**(t[i]-1)

#             J[i][0] = -np.exp(x[1]*t[i])
#             J[i][1] = -t[i]*x[0]*np.exp(x[1]*t[i])
        
    return J
```

```python
def gauss_newton(data, f, time, x_init, log = 0):
    '''
    data : given y
    f : f or log_f function to pass
    time : time taken 
    x_init : starting vector (2 x 1)
    log: T/F : boolean
    
    Returns: datapoints as [P0, r]
    '''
    
    k = 0    
    ans = []
    s = np.ones(1)
    e_mach = 2**-52
    
    print("Gauss Newton sequence of iterations: \n")
    t = PrettyTable(['k', 'x.T', 'r2'])
    
    while(any(s) > e_mach):
                
        ans.append(x_init)
        
        res = r(data, f, time, x_init)
        r2 = np.linalg.norm(res, 2) ** 2
        
        t.add_row([k, x_init, r2])
        
        Jx = J(time, x_init, log)
        s = np.linalg.lstsq(Jx, -res, rcond=None)[0]        
        
        x1 = x_init + s
        
        s = abs(np.round(s, 6))
        x_init = x1
        k = k + 1
    
    print(t)
    return np.array(ans)
```

```python
# data = np.array([2.0, 0.7, 0.3, 0.1])
```

```python
x_init = np.array([0.4, 1.7], dtype = float)
ti = np.array([i for i in range(len(data))], dtype = float)
```

```python
ans = gauss_newton(data, f, ti, x_init, 0)
```

```python
def population(p):
    '''
    p matrix: 1st column - p0, 2nd column - r
    p_t = r * p_t-1
    '''
    p1 = p[:, 0]*p[:, 1]
    return p1
```

# Exercise 3



Runge's function is given by

$$ f(t) = 1/(1+25t^2).$$

# (a)
Interpolate Runge's function on $[-1,1]$ by a polynomial $p_k$ of degree $k$ for $k = 5,10,15$. Use equally spaced interpolation points. Draw your interpolating polynomials together with $f$.

```python
def runge_interpolate(degree, p):
    '''
    degree: k : [5, 10, 15]
    p: number of points (ex: 100)
    '''
    
    points = np.linspace(-1, 1, p+1)
    degree_pk_dict = {}
    
    for k in degree:
        degree_pk_dict['d'+str(k)] = np.linspace(-1, 1, k+1)
        degree_pk_dict['pk'+str(k)] = 1/(1+25*degree_pk_dict['d'+str(k)]**2)
        pk_fit = np.polyfit(degree_pk_dict['d'+str(k)], degree_pk_dict['pk'+str(k)], k)
        degree_pk_dict['ft_k'+str(k)] = np.polyval(pk_fit, points)
    
    ft_interpolate = 1/(1+25*points**2)
    
    return degree_pk_dict, points, ft_interpolate

def plot_interpolate(obj, k, title_, xlab):
    
    ls = ['-', '--', '-.', ':', '-', '--', '-.', ':']
    cs = ['lightcoral', 'saddlebrown', 'goldenrod', 'limegreen', 'turquoise', 'slategray', 'mediumorchid', 'coral']
    degree_pk_dict, points, ft_interpolate = obj
    
    for j, i in enumerate(k):
        plt.plot(points, degree_pk_dict['ft_k'+str(i)], linestyle = ls[j], 
                 label = "{}th degree polynomial".format(len(degree_pk_dict['d'+str(i)])-1),
                 color = cs[j])
    
    plt.plot(points, ft_interpolate, label = "f(t)")
    plt.plot(degree_pk_dict['d5'], degree_pk_dict['pk5'], 'D', markersize = 8,
             label = "{} equally spaced datapoints".format(len(degree_pk_dict['d5'])-1))
    
    plt.ylim(-0.7, 2.75)
    plt.title(title_, fontsize=14)
    plt.xlabel("Range [-1, 1]", fontsize=14)
    plt.ylabel(xlab, fontsize=14)
    plt.legend()
    plt.grid(0)
    plt.show()
```

```python
kr = [3, 5, 10, 15, 20]
robj = runge_interpolate(kr, 500)
plot_interpolate(robj, kr, "Interpolants of Runge's function at equally spaced points", r'$f(t)$')
```

# (b)
Repeat (a) to find polynomials $q_k$. This time use the Chebyshev points as interpolation points. They are given by

$$t_i = \cos \left( \frac{(2i-1)\pi}{2k+2} \right),\, \,i = 1,\ldots,k+1.$$

```python
def cheb_func(p):
    
    points = np.linspace(-1, 1, p+1)
    
    for i in range(len(points)):
        points[i] = np.cos( ((2*(i+1) -1)*np.pi)/(2*p + 2) )
    
    return points
```

```python
def chebyshev_interpolate(degree, p):
    '''
    degree: k : [5, 10, 15]
    p: number of points (ex: 100)
    '''
    
    points = cheb_func(p)
    
    degree_pk_dict = {}
    
    for k in degree:
        degree_pk_dict['d'+str(k)] = cheb_func(k)
        degree_pk_dict['pk'+str(k)] = 1/(1+25*degree_pk_dict['d'+str(k)]**2)
        pk_fit = np.polyfit(degree_pk_dict['d'+str(k)], degree_pk_dict['pk'+str(k)], k)
        degree_pk_dict['ft_k'+str(k)] = np.polyval(pk_fit, points)
    
    ft_interpolate = 1/(1+25*points**2)
    
    return degree_pk_dict, points, ft_interpolate
```

```python
kc = [3, 5, 10, 15, 20, 25, 50, 100]
cobj = chebyshev_interpolate(kc, 500)
plot_interpolate(cobj, kc, "Interpolants of Runge's function at Chebyshev points", r'$q_k$')
```

# (c)
Verify numerically that the maximum of the difference of $f$ and $q_k$ in the interval $[-1,1]$ decreases with increasing $k$ by trying various values of $k$.

$$
\underset{x \in[-1, 1]}{\max}|f(x) - q_k(x)|
$$

```python
def max_diff(k, obj):
    ans = []
    for i in k:
        md = max(abs( obj[2] -obj[0]['ft_k'+str(i)] ))
        print("Max abs difference between f and q_k for k {} is {}".format(len(obj[0]['d'+str(i)])-1, md))
        ans.append(md)

    fig, ax = plt.subplots()
    ax.scatter(k, ans, marker = "X", s = 200)
    
    for i, txt in enumerate(k):
        ax.annotate('(k = '+str(txt)+')', (k[i]+1, ans[i]))
    
    plt.xlabel("k")
    plt.ylabel("max abs diff")
    plt.show()
```

It can be clearly seen from these results that as k increases the maximum difference between f and $q_k$ decreases
